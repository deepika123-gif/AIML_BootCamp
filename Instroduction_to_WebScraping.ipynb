<h2><p><b>*Requests Module:
</b>The requests library is used for making HTTP requests to a specific URL and returns the response. Python requests provide inbuilt functionalities for managing both the request and response.
</p></h2>


```python
#import liberes
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
url="https://www.flipkart.com/search?q=t+shirts&as=on&as-show=on&otracker=AS_Query_TrendingAutoSuggest_4_0_na_na_na&otracker1=AS_Query_TrendingAutoSuggest_4_0_na_na_na&as-pos=4&as-type=TRENDING&suggestionId=t+shirts&requestId=c15f8b75-4a1c-4409-a216-7501ee0ddb2d"
```

```python
response=requests.get(url)
```

```python
print(response.status_code )
print(response.content)
```

<h3><p> <b> Explanation:</b>
<p>*<u>
requests.get(url):</u> Sends a GET request to the given URL.</p><p>*
<u>response.status_code:</u> Returns HTTP status code (200 = success).</p><p>*
<u>response.content:</u> Returns the raw HTML of the page in bytes.</p>
</h3>

<h2>   <p><b>  *Parsing HTML with BeautifulSoup:</b>
Once the raw HTML is fetched, the next step is to parse it into a readable structure. Thats where BeautifulSoup comes in. It helps convert the raw HTML into a searchable tree of elements.
</p></h2>


```python
import requests
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(response.content, 'html.parser')

print(soup.prettify())
```

<h3><p> <b> Explanation:</b>
<p>*<u>
BeautifulSoup(html, parser):</u>Converts HTML into a searchable object. 'html.parser' is the built-in parser. </p><p>*
<u>soup.prettify(): </u>Formats the HTML nicely for easier reading.
</p></h3>

<h2>   <p><b>  *Extracting Content by Tag and Class:</b>
Once we have parsed the HTML using BeautifulSoup, the next step is to locate and extract specific content from the page</p></h2>


```python
products=soup.find_all("div",class_="_1sdMkc LFEi7Z")
```

```python
print(len(products))
```

```python
products
```

```python
print(soup.prettify())
```

```python
t_shirt=[]
```

```python
for i in products:
    t_shirt_name=i.div
    if t_shirt_name:
        title=t_shirt_name.get_text(strip=True)
    price=i.find("div",class_="Nx9bqj")
    if price:
        price=price.get_text(strip=True)
        
    t_shirt.append({
        'Title': title,
        "Price":price,
    })
```

```python
print(t_shirt)
```

```python
df=pd.DataFrame(t_shirt)
df
```

```python
df.to_csv("flipkart.csv")
```

<h1>-->  Popular Tools for Web Scraping
</h1></div>
<h2><p> There are several tools and libraries available that make web scraping easier and more efficient. Some are lightweight for beginners while others are built for large-scale data extraction:
   </p></h2>

   
   <h2><p><ul><li><b> BeautifulSoup:</b>
   BeautifulSoup is a beginner-friendly Python library used to parse HTML and XML documents. It allows us to navigate the page structure and extract specific elements using tags and classes.
   </li></p><p>
   <li><b> Requests:
    </b>
Requests is used with BeautifulSoup as it helps to send HTTP requests to websites and fetch the HTML content of web pages.
 </li></p><p>
<li>  <b> Selenium:</b>
Selenium is a web automation tool that can control a browser like a real user. Its useful for scraping websites that use JavaScript to load content such as infinite scrolling or dropdown menus.

 </li>  </p><p> <li><b> Scrapy:</b>
Scrapy is an advanced Python framework built for web scraping. It supports features like crawling, handling requests/responses, managing pipelines and storing scraped data efficiently.
</li></p><p>  <li><b> Playwright:</b>
Playwright is a newer alternative to Selenium, it supports modern web standards and provides better performance for scraping dynamic content with headless browser control.
</li></p>


   </ul></p></h2>

<h1>--> Challenges to Web Scraping

</h1></div>
<h2><p>
Although web scraping is useful, it comes with several challenges that can make the process difficult:
<h2><p><ul><li> <b>
Website Structure Changes:</b> Websites sometimes update their design and code which can break scraping scripts that rely on specific HTML elements. Scrapers need regular maintenance to keep up with these changes.</li><p>
<li><b>
Anti-Scraping Technologies:</b> Many websites use measures like IP blocking, CAPTCHA or dynamic content loading to prevent automated scraping.</li></p><p><li><b>
Data Storage and Management:</b> Large-scale scraping generates huge volumes of data. Efficiently storing organizing and processing this data requires good infrastructure and planning.</li></p><p><li><b>
Ensuring Data Quality:</b> Extracted data might be incomplete, duplicated or outdated. Cleaning and validating data to maintain accuracy is an important but challenging step.</li></p><p><li><b>
Legal Risks:</b> As discussed earlier, scraping without permission or violating terms of service can lead to legal consequences or blocked access.</li></p><p><li><b>
Performance and Speed:</b> Balancing fast data extraction while avoiding detection or server overload requires careful handling of request rates and scraping strategies.
   </li></p> 
</p></h2>
