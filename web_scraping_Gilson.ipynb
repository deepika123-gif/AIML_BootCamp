<h1><p><b>Scraping Website (Gilson)

<h2><p><b>Importing Libraries


```python
import pandas as pd
import requests 
from bs4 import BeautifulSoup
```

<h2><p><b>Requesting url


```python
url="https://www.gilson.com/default/shop-products/pipettes.html?srsltid=AfmBOoqOV3Sdidb7prT6f91fn9c_JRv_uo8cpJNV_r-EaKW_5kikp4hx"
```

```python
response=requests.get(url)
response
```

<h2><p><b>Trying to Fetch Webpage


```python

if response.status_code == 200:
    print("Successfully fetched the page!")
else:
    print(f"Failed to retrieve the page. Status code:{response.status_code}")
```

<h2><p><b>Accessing the Code


```python
soup = BeautifulSoup(response.content, "html.parser")
soup
```

<h2><p><b>Prettify the code

```python
print(soup.prettify())
```

<h2><p><b>Finding All Division Class That Contains the Details



```python
pipette = soup.find_all('div', class_="product details product-item-details")
pipette
```

<h2><p><b>Finding The Number of Division Class



```python
print(len(pipette))
```

<h2><p><b>Creating Empty List

```python
pipette_data=[]
```

<h2><p><b>Accessing required Information



```python
for p in pipette:
    
    product_element = p.find('div', class_="sku-box")
    product = product_element.get_text(strip=True) if product_element else "N/A"

    title_element = p.find('a', class_="product-item-link")
    title = title_element.get_text(strip=True) if title_element else "N/A"

    price_element = p.find('span', class_="price")
    price = price_element.get_text(strip=True) if price_element else "N/A"

    pipette_data.append({
      "Product ID": product,
      "Product Name": title,
      "Price": price,
    })
```

```python
pipette_data
```

<h2><p><b>Converting into DataFrame



```python
df = pd.DataFrame(pipette_data)
df
```

<h2><p><b>Saving to CSV file



```python
df.to_csv("GilsonPipettes.csv")
```
