<h1><p><b>Scraping Website PubMed(National Library of Medicine)

<h2><p><b>Importing Libraries



```python
import pandas as pd
import requests 
from bs4 import BeautifulSoup
```

<h2><p><b>Requesting url



```python
url="https://pubmed.ncbi.nlm.nih.gov/?term=asprin&filter=datesearch.y_1"
```

```python
response=requests.get(url)
response
```

<h2><p><b>Trying to Fetch Webpage



```python

if response.status_code == 200:
    print("Successfully fetched the page!")
else:
    print(f"Failed to retrieve the page. Status code:{response.status_code}")
```

<h2><p><b>Accessing the Code



```python
soup = BeautifulSoup(response.content, 'html.parser')
soup
```

<h2><p><b>Prettify the code

```python
print(soup.prettify())
```

<h2><p><b>Finding All Division Class That Contains the Details



```python
data=soup.find_all("article" ,class_='full-docsum')
data
```

<h2><p><b>Finding The Number of Division Class



```python
print(len(data))
```

<h2><p><b>Printing Those Division Class



```python
data
```

<h2><p><b>create empty list

```python
medicine=[]
```

<h2><p><b>Accessing required Information



```python
for i in data:
    medicine_name = i.find("a",class_="docsum-title")
    PMID = i.find("span", class_="docsum-pmid")
    description = i.find("div", class_="full-view-snippet")

    if medicine_name and PMID and description:
        medicine_name=medicine_name.get_text(strip=True)
        PMID=PMID.get_text(strip=True)
        description=description.get_text(strip=True)
    

    medicine.append({
        "Name":medicine_name,
        "PMID":PMID,
        "Description":description,
    })
```

```python
medicine
```

<h2><p><b>Converting into DataFrame



```python
df=pd.DataFrame(medicine)
df
```

<h2><p><b>Saving to CSV file



```python
df.to_csv("medicine data.csv")
df
```
