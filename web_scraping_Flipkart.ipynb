<h1><p><b>Scraping Flipkart-->Laptops Website

<h1><p><b>Importing Libraries

```python
import requests                 
from bs4 import BeautifulSoup    
import pandas as pd  
```

<h1><p><b>Requesting url


```python
url="https://www.flipkart.com/search?q=laptop&otracker=AS_Query_HistoryAutoSuggest_3_0&otracker1=AS_Query_HistoryAutoSuggest_3_0&marketplace=FLIPKART&as-show=on&as=off&as-pos=3&as-type=HISTORY"
```

<h1><p><b>Trying to Fetch Webpage


```python
response = requests.get(url)
response
```

```python
if response.status_code == 200:
    print("Successfully fetched the page!")
else:
    print(f"Failed to retrieve the page. Status code:{response.status_code}")
```

<h1><p><b>Adding Headers


```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
 }
```

<h1><p><b>Getting Response


```python
response = requests.get(url, headers=headers)
response
```

```python
if response.status_code == 200:
    print("Successfully fetched the page!")
else:
    print(f"Failed to retrieve the page. Status code:{response.status_code}")
```

<h1><p><b>Accessing the Code


```python
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
soup
```

<h1><p><b>Prettify the Code


```python
print(soup.prettify())
```

<h1><p><b>Finding All Division Class That Contains the Details


```python
laptop=soup.find_all("div" ,class_='tUxRFH')
```

```python
laptop
```

<h1><p><b>Finding The Number of Division Class


```python
print(f"Number of laptop found: {len(laptop)}")
```

<h1><p><b>Printing Those Division Class


```python
laptop
```

<h1><p><b>Creating Empty List


```python
laptop_data=[]
```

<h1><p><b>Accessing required Information


```python
for i in laptop:
    name= i.find("div", class_="KzDlHZ")
    price=i.find("div",class_="Nx9bqj _4b5DiR")
    rating=i.find("div",class_="XQDdHH")
    
    if rating:
        ratings=rating.get_text(strip=True)

    laptop_data.append ( {
    "Name":name,
    "Author":price,
    "Rating":ratings
    
    
    }) 
```

```python
laptop_data
```

<h1><p><b>Converting into DataFrame


```python
df=pd.DataFrame(laptop_data)
```

```python
df
```

<h1><p><b>Saving to CSV file


```python
df.to_csv("Flipkart-Laptop.csv")
```
