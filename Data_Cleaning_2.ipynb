```python
# Convert the object datatype into numeric datatype
 # importing the pandas library for mathematic computation
import pandas as pd


df['female_male_ratio']=pd.to_numeric(df['female_male_ratio'],
errors='coerce')    #used for-any non numeric values(like text) will be converted to NAN(missing values)
```

```python
# Calculate the mean of the 'female_male_ratio' column
mean_female_male_ratio=df['female_male_ratio'].mean()
mean_female_male_ratio
```

```python
df1=df.copy()
```

```python
#filling missing values in 'female_male_ratio' with the mean value 

df1['female_male_ratio'].fillna(mean_female_male_ratio)  
```

```python
#for removing the decimal part from the data use floor division

mean_female_male_ratio//1
```
