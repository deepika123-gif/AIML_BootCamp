<h1><p><b>Scraping-->Flipkart-->Laptops(laptop name,price,rating)

```python
import pandas as pd 
import requests 
from bs4 import BeautifulSoup
```

```python
url="https://www.flipkart.com/search?q=laptops&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"
```

```python
response=requests.get(url)
response
```

```python
if response.status_code == 200:
    print("Successfully fetched the page!")
else:
    print(f"Failed to retrieve the page. Status code:{response.status_code}")
```

```python
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
soup
```

```python
print(soup.prettify())
```

```python
product=soup.find_all("div",class_="tUxRFH")
```

```python
print(len(product))
```

```python
product
```

```python
laptop=[]
```

```python
for i in product:

    laptop_name=i.find("div",class_="KzDlHZ")
    if laptop_name:
        title=laptop_name.get_text(strip=True)

    price=i.find("div",class_="Nx9bqj _4b5DiR")
    if price:
        price=price.get_text(strip=True)

    rating=i.find("div",class_="XQDdHH")
    if rating:
      Rating=rating.get_text(strip=True)

        
    laptop.append({
        'Title': title        ,
        "Price":price,
        "Rating":Rating,

    })
```

```python
print(laptop)
```

```python
df=pd.DataFrame(laptop)
df
```

<h1><p><b>Visualization


```python
import matplotlib.pyplot as plt
```

```python
#bar plot

plt.bar(df["Title"],df["Price"])
plt.title("Bar plot")
plt.xlabel("Title")
plt.ylabel("Price")
plt.show()
```

```python
#scatter plot

plt.scatter(df["Title"],df["Price"])
plt.title("sample plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
#pie chart

from collections import Counter
counts=Counter(df["Rating"])
label=list(counts.keys())
size=list(counts.values())

plt.pie(size,labels=label,autopct='%1.1f%%')
plt.title("sample plot")
plt.xlabel("subject")
plt.ylabel("marks")
plt.show()
```
