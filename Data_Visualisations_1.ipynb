<h1><ul><li>Bar Plots</li></ul></h1>

```python
#example data

abc=['Deepika','Neha','Suhana','Kirti','Rezal']
values=[100,200,150,280,300]

#Bar plot

#import library
import matplotlib.pyplot as plt

plt.bar(abc,values,color='green',width=0.8)
plt.xlabel('names')
plt.ylabel('marks')
plt.title('marks')



```

<h1><ul><li>Scatter Plot</li></ul></h1>

```python
x=['hindi','english','math','punjabi','computer science']
y=[30,53,60,75,50]
z=[10,200,40,30,100]

plt.scatter(x,y,color='blue',marker='s')

plt.scatter(x,z,color='green',marker='o')

plt.xlabel('subjects')
plt.ylabel('marks')
plt.title('marks')
```

<h1><ul><li>Line Plot</li></ull></h1>

```python
x=[1,2,3,4,5,6,7,8]
a=[10,30,70,40,50,20,35,15]

plt.plot(x,a, label='nvidia', color='black')
```

<h1><p>#Company Sales Comparison</p></h1>

```python
import numpy as np

x=[1,2,3,4,5,6]                             #years
a=[10,20,40,50,90,30]                      #sales
b=[5,10,50,20,15,50]                       #sales
c=[100,200,50,80,400,200]

plt.plot(x,a, label='oopo A15s', color='blue')
plt.plot(x,b, label='vivo', color='red')
plt.plot(x,c, label='apple', color='green')

plt.xlabel('years')
plt.ylabel('sales')
plt.title('company sales comparison')
plt.legend()
plt.show()
```

<h1><ul><li>Histogram</li></ul></h1>

```python
import seaborn  as sns

iris=sns.load_dataset('iris')
```

```python
iris.head()
iris.info()
iris.describe()
```

```python
sns.histplot(data=iris, x='petal_length', bins=100, kde=True,color='red')
```

```python
sns.kdeplot(data=iris, x='petal_length', hue='species', color='green',fill=True,alpha=0.5)
```

```python
column=['species']

#count of each species
iris['species'].value_counts()
```

```python
#skewedness check
sns.countplot(x='species', data=iris)
```

<h2><p>Pairplot</p></h2>

```python
sns.pairplot(iris, markers=["o","s","p"], hue='species',palette='husl')
```
