<h1><p><b>Scraping Website (Flipkart-->Phones)

<h2><p><b>Importing Libraries

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

<h2><p><b>Requesting url


```python
url="https://www.flipkart.com/search?q=mobile&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"
```

```python
response=requests.get(url)
response
```

<h2><p><b>Trying to Fetch Webpage


```python

if response.status_code == 200:
    print("Successfully fetched the page!")
else:
    print(f"Failed to retrieve the page. Status code:{response.status_code}")
```

<h2><p><b>Adding Headers


```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
 }
```

<h2><p><b>Getting Response


```python
response = requests.get(url, headers=headers)
response
```

```python

if response.status_code == 200:
    print("Successfully fetched the page!")
else:
    print(f"Failed to retrieve the page. Status code:{response.status_code}")
```

<h2><p><b>Accessing the Code


```python
soup = BeautifulSoup(response.content, 'html.parser')
soup
```

<h2><p><b>Prettify the Code


```python
print(soup.prettify())
```

```python
phones=soup.find_all("div" ,class_='tUxRFH')
phones
```

<h2><p><b>Finding The Number of Division Class


```python
print(f"Number of phones found: {len(phones)}")
```

<h2><p><b>Printing Those Division Class


```python
phones
```

<h2><p><b>Creating Empty List


```python
phone_data=[]
```

<h2><p><b>Accessing required Information


```python
for i in phones:
    name = i.find("div", class_="KzDlHZ")
    price = i.find("div", class_="Nx9bqj _4b5DiR")
    rating = i.find("div", class_="XQDdHH")
    storage = i.find("li", class_="J+igdf")

    phone_data.append({
        "Name": name.get_text(strip=True) if name else "N/A",
        "Price": price.get_text(strip=True) if price else "N/A",
        "Rating": rating.get_text(strip=True) if rating else "N/A",
        "Storage": storage.get_text(strip=True) if storage else "N/A",
    })
```

```python
phone_data
```

<h2><p><b>Converting into DataFrame


```python
df=pd.DataFrame(phone_data)
df
```

<h2><p><b>Saving to CSV file


```python
df.to_csv(" phones Data.csv")
```
