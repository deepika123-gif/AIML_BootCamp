<h2><p>Section 4-Fetch live data

```python
import time
ticker = yf.Ticker("AAPL")

while True:
    data = ticker.history(period = "1d", interval="1m")
    current_price =data['Close'].iloc[-1]
    current_time = data.index[-1]
    print(f"{current_time} Current AAPL prize: ${current_price}")
    time.sleep(60)
```
