```python
#drop(delete) missing value row
df_listwise_deletion=df.dropna()
```

```python
#calculate the mean of the 'Age'column

mean_quantity=df['Age'].mean()
mean_quantity

# create a copy of the DataFrame to avoid modifying the original DataFram
df1=df.copy()
```

```python
# fill missing values in 'Quantity' with the mean value

df1['Age'].fillna(mean_quantity)
```

```python
mean_quantity//1
```

```python
# round the mean value to 0 decimal places
round(mean_quantity)
```

```python
 # fill missing values in 'Quantity' with the rounded mean value
df1['Age'].fillna(round(mean_quantity), inplace=True)
```

```python
df['Pclass'].value_counts().plot(kind='bar',color='blue')
plt.xlabel('passenger Class')
plt.ylabel('Count')
plt.title('number of passengers by Class')
plt.show()
```

<h2><p><ul><li>Scatter Plot</li></ul></p></h2>

```python
plt.scatter(df['Age'],df['Fare'],alpha=0.7,color='red')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter pot of Age and Fare')
plt.show()
```

<h2><p><ul><li>Line Plot
</li></ul></p></h2>

```python
plt.plot(df['Age'][:50],label='Age')
plt.plot(df['Fare'][:50], label='Fare')
plt.xlabel('Passenger Index')
plt.ylabel('value')
plt.title('Line plot of Age and Fare')
plt.legend()
plt.show()
```

<h2><p><ul><li>Histogram Plot</li></ul></p></h2>

```python
plt.hist(df['Age'].dropna(),bins=30,color='purple',edgecolor='red')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution')
plt.show()
```
