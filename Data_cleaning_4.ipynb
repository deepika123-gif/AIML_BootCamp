```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
#creating syntheitc data and loading it

data={
    "patient_id":[1,2,3,4,5,6,7,8],
    "name":["neha","Deepika","chetan","dheeraj","isha","guri","rajan","komal"],
    "age":[34,np.nan,32,88,45,33,np.nan,46],
    "gender":['f','f','m','m','f','m','m','f'],
    "blood_pressure":['120/80',"none",'140/90','abc','130/85','150/95','120/80','135/85'  ],
    "diagonsis":["hypertension","disbetes","hypertension","none","diabetes","HYPERTENSION","pre-dibetes","hypertension"],
    "date_of_visit":["2023-01-10","2023-03-15","15/02/2023","2023/04/18","2023//05/20","invails_date","2023/04/2","2023/03/04"]
}
```

```python
df=pd.DataFrame(data)
```

```python
df
```

```python
df['diagonsis']=df['diagonsis'].str.lower().str.strip()
```

```python
df
```

```python
df['diagonsis'].replace(["none","null",""],np.nan,inplace=True)
df
```

```python
df['age']=df['age'].fillna(df['age'].median())
df
```

```python
df
```

```python
df['date_of_visit']=pd.to_datetime(df['date_of_visit'],errors='coerce')
```

```python
df
```

```python
#remove duplicated value

df.drop_duplicates(inplace=True)
```

```python
df.dropna(subset=['blood_pressure','diagonsis' ,'date_of_visit'],inplace=True)
```
