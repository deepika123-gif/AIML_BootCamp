```python
import pandas as pd
import requests 
from bs4 import BeautifulSoup
```

```python
url="https://books.toscrape.com/"
```

```python
response = requests.get(url)
response
```

```python
if response.status_code == 200:
    print("Successfully fetched the page!")
else:
    print(f"Failed to retrieve the page. Status code:{response.status_code}")
```

```python
soup = BeautifulSoup(response.content, 'html.parser')
soup
```

```python
print(soup.prettify())
```

```python
books=soup.find_all("li",class_="col-xs-6 col-sm-4 col-md-3 col-lg-3")
books
```

```python
print(f"Number of qoutes found: {len(books)}")
```

```python
books
```

```python
book_data=[]
```

```python
for i in books:
    book_name = i.find("h3").find("a")
    book_price = i.find("p", class_="price_color")
    book_rating = i.find("p", class_="star-rating")

    book_name = book_name.get("title") if book_name else "N/A"

    if book_price:
        price_text = book_price.get_text(strip=True) 
        price_value = float(price_text.replace("£", ""))
        book_price = f"₹{price_value * 100:.2f}"   
    else:
        book_price = "N/A"

    if book_rating:
        classes = book_rating.get("class")
        rating_word = classes[1] if len(classes) > 1 else "N/A"
        ratings = ["Zero","One","Two","Three","Four","Five"]
        book_rating = ratings.index(rating_word) if rating_word in ratings else "N/A"
    else:
        book_rating = "N/A"

    book_data.append({
        "Book Name": book_name,
        "Book Price": book_price,
        "Book Rating": book_rating,
    })
```

```python
book_data
```

```python
df=pd.DataFrame(book_data)
```

```python
df
```

```python
df.to_csv(" books.csv")
```
