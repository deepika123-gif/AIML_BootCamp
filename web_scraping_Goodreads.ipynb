<h1><p><b>Scraping Goodreads Website



<h2><p><b>Importing Libraries

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

<h2><p><b>Requesting url

```python
url = 'https://www.goodreads.com/quotes'
```

```python
response = requests.get(url)
response
```

<h2><p><b>Trying to Fetch Webpage



```python

if response.status_code == 200:
    print("Successfully fetched the page!")
else:
    print(f"Failed to retrieve the page. Status code:{response.status_code}")
```

<h2><p><b>Adding Headers



```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
 }
```

<h2><p><b>Getting Response

```python
response = requests.get(url, headers=headers)
response
```

```python

if response.status_code == 200:
    print("Successfully fetched the page!")
else:
    print(f"Failed to retrieve the page. Status code:{response.status_code}")
```

<h2><p><b>Accessing the Code



```python
soup = BeautifulSoup(response.content, 'html.parser')
soup
```

<h2><p><b>Prettify the Code

```python
print(soup.prettify())
```

<h2><p><b>Finding All Division Class That Contains the Details



```python
quotes=soup.find_all("div" ,class_='quote')
quotes
```

<h2><p><b>Finding The Number of Division Class



```python
print(f"Number of qoutes found: {len(quotes)}")
```

<h2><p><b>Printing Those Division Class



```python
quotes
```

<h2><p><b>Creating Empty List

```python
quotes_data=[]
```

<h2><p><b>Accessing required Information



```python
for i in quotes:
    title= i.find("div", class_="quoteText").get_text(strip=True)
    Author=i.find("span",class_="authorOrTitle").get_text(strip=True)


    quotes_data.append ( {
    "Quote":title,
    "Author":Author
    
    
    }) 
```

```python
quotes_data
```

<h2><p><b>Converting into DataFrame



```python
df=pd.DataFrame(quotes_data)
```

```python
df
```

<h2><p><b>Saving to CSV file



```python
df.to_csv(" Quotes.csv")
```
