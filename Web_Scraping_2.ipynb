<h1><p><b>Scraping Quotes to Scrape Website</b></p></h1> 

<h1><p><b>Importing Libraries


```python
import requests                 
from bs4 import BeautifulSoup    
import pandas as pd  
```

<h1><p><b>Requesting url


```python
url="https://quotes.toscrape.com/"
```

```python
response = requests.get(url)
response
```

<h1><p><b>Trying to Fetch Webpage


```python
if response.status_code == 200:
    print("Successfully fetched the page!")
else:
    print(f"Failed to retrieve the page. Status code:{response.status_code}")
```

<h1><p><b>Accessing the Code

```python
soup = BeautifulSoup(response.content, 'html.parser')
soup
```

<h1><p><b>Prettify the Code


```python
print(soup.prettify())
```

<h1><p><b>Finding All Division Class That Contains the Details


```python
quotes=soup.find_all("div" ,class_='quote')
```

```python
quotes
```

<h1><p><b>Finding The Number of Division Class


```python
print(f"Number of qoutes found: {len(quotes)}")
```

<h1><p><b>Printing Those Division Class


```python
quotes
```

<h1><p><b>Creating Empty List


```python
quotes_data=[]
```

<h1><p><b>Accessing required Information


```python
for i in quotes:
    title= i.find("span", class_="text").get_text(strip=True)
    Author=i.find("small",class_="author").get_text(strip=True)


    quotes_data.append ( {
    "Quote":title,
    "Author":Author
    
    
    }) 
```

```python
quotes_data
```

<h1><p><b>Converting into DataFrame


```python
df=pd.DataFrame(quotes_data)
```

```python
df
```

<h1><p><b>Saving to CSV file


```python
df.to_csv("Popular Quotes.csv")
```
