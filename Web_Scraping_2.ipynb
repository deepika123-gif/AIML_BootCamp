<h1><p><b>scraping Flipkart WebSite</b></p></h1>

```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python
url="https://www.flipkart.com/search?q=t+shirts&as=on&as-show=on&otracker=AS_Query_TrendingAutoSuggest_4_0_na_na_na&otracker1=AS_Query_TrendingAutoSuggest_4_0_na_na_na&as-pos=4&as-type=TRENDING&suggestionId=t+shirts&requestId=c15f8b75-4a1c-4409-a216-7501ee0ddb2d"
```

```python
response=requests.get(url)
```

```python
response = requests.get(url)
response
```

```python
if response.status_code == 200:
    print("Successfully fetched the page!")
else:
    print(f"Failed to retrieve the page. Status code:{response.status_code}")
```

```python
soup2 = BeautifulSoup(response.content, 'html.parser')
```

```python
soup2
```

```python
print(soup2.prettify())
```

```python
products=soup2.find_all("div",class_="_1sdMkc LFEi7Z")
```

```python
print(len(products))
```

```python
products
```

```python
print(soup2.prettify())
```

```python
t_shirt=[]
```

```python
for i in products:
    t_shirt_name=i.div
    if t_shirt_name:
        title=t_shirt_name.get_text(strip=True)
    price=i.find("div",class_="Nx9bqj")
    if price:
        price=price.get_text(strip=True)
        
    t_shirt.append({
        'Title': title,
        "Price":price,
    })
```

```python
print(t_shirt)
```

```python
df=pd.DataFrame(t_shirt)
```

```python
df
```

```python
df.to_csv("flipkart.csv")
```
